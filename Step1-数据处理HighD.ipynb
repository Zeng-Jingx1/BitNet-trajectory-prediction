```python

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']   #显示中文
plt.rcParams['axes.unicode_minus']=False       #显示负号
plt.rcParams['font.family'] = 'Times New Roman'

import pandas as pd 
import numpy as np
from tqdm import tqdm
import os
```

```python
tracks = pd.read_csv('./HighD/01_tracks.csv')
```

```python
tracks
```

```python
json_files = json_files = tracks['id'].unique()
json_files = json_files[:300]
```

```python


data_container = []
for p in tqdm(json_files):
    demo =  tracks[tracks['id']==p]
    demo = demo[['x','y']]
    data_container.append(demo)
import pickle 
# 保存
with open('datahd.pkl', 'wb') as f:
    pickle.dump(data_container, f)
```
